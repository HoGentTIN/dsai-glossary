# Data Science & AI - afbeeldingen

Dit bestand bevat de broncode voor de afbeeldingen ter illustratie van de definities in deze woordenlijst.

```python
# Package imports for data science
import numpy as np
import pandas as pd
import scipy.stats as stats
from pandas.api.types import CategoricalDtype
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.graphics.mosaicplot import mosaic
```

## Normale verdeling



```python
# Parameters
m = 25
s = 2.5

# X- en Y-coördinaten van de te plotten punten
x = np.linspace(start=m - 4 * s, stop=m + 4 * s, num=101)
y = stats.norm.pdf(x, loc=m, scale=s)


# Kansdichtheid
plt.plot(x, y);
# Gemiddelde
plt.axvline(x=m, color='orange');
# ±1x stdev
plt.axvline(x=m - s, color='blue', linestyle=':');
plt.axvline(x=m + s, color='blue', linestyle=':');
# ±2x stdev
plt.axvline(x=m - 2 * s, color='green', linestyle=':');
plt.axvline(x=m + 2 * s, color='green', linestyle=':');
# ±3x stdev
plt.axvline(x=m - 3 * s, color='red', linestyle=':');
plt.axvline(x=m + 3 * s, color='red', linestyle=':');

varname = 'Z' if (m == 0 and s == 1) else 'X'
plt.title(f'{varname} ~ Nor(μ={m}, σ={s})');
```
